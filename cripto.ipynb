# Programme 1

- parcourir la chaine __message__ de gauche à droite en comptant le nombre d’occurrences successives de chaque caractère de la chaine,
- construire la chaine __chiffre__ avec ce nombre suivi du caractère en question.

## Question 

Ecrire un programme permettant de saisir la chaine __message__ qui doit être non vide et formée uniquement par des lettres alphabétiques, puis construire et afficher la chaine __chiffre__ selon le principe décrit précédemment.

__Exemple__ :
    Si __message__ = aaaFyBssssssssssssazz alors la chaine __chiffre__ qui sera affichée est 3a1F1y1B12s1a2z

Pour vous aider, vous pouver utiliser les fonctions suivantes : len, range, str

Vous trouverez ci-dessous des exemples d'utilisations :

```python
message = "aaaFyBssssssssssssazz"
len(message)
```

```python
cpt = 1
for i in range(len(message)-1):
    print(message[i], i)
    cpt += 1
```

```python
chiffre = str(cpt)
print(chiffre)
```

```python
def chiffrer(message):
    message = "aaaFyBssssssssssssazz"
    chiffre = ""
    cpt = 1
    for i in range(len(message)-1):
        if message [i] == message[i+1]:
            cpt += 1
        else : 
            chiffre += str(cpt) + message[i]
            cpt=1
    chiffre += str(cpt) + message [len(message)-1]
    return chiffre
        
print(chiffrer("aaaFyBssssssssssssazz"))
```

# Programme 2

On se propose d’écrire un programme qui permet de saisir et de chiffrer un mot __message__ non vide, composé uniquement par des lettres majuscules et d’afficher le mot chiffré __chiffre__.

La méthode de chiffrement est la suivante :

- Pour chaque lettre du mot __message__, déterminer son nombre d’occurrence (apparition) *n* dans le mot. 

- Déterminer *indice* qui est égal à *2 x n* si *n* est impair et égal à *n / 2* si *n* est pair. 

- Remplacer chaque lettre par la lettre avec l'indice *indice* qui la suit dans l’intervalle de l’alphabet __ABCDEFGHIJKLMNOPQRSTUVWXYZ__. 

- Pour les dernières lettres, on revient au début (opération modulo).

- Par exemple, si *indice = 3*, on remplacera __A__ par __D__, __B__ par __E__, __C__ par __F__, puis __Y__ par __B__ et __Z__ par __C__.

Par exemple si le __message__ est HAPPY

Le chiffré est : JCQQA

## Question

Ecrire un programme qui permet de saisir un mot __messsage__ non vide et composé uniquement par des lettres majuscules, puis afficher le mot chiffré __chiffre__ selon le principe décrit ci-dessus.

Pour vous aider, vous pouver utiliser les fonctions suivantes : count, find

Vous trouverez ci-dessous des exemples d'utilisations :

```python
message = "BONJOUR"
for lettre in message :
        occurence = message.count(lettre)
        print(lettre, occurence)
```

```python
alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
lettre = 'V'
ordre = alpha.find(lettre)
print(ordre)
```

```python
def chiffrer(message):
    chiffre = ""
    alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    for lettre in message :
        occurence = message.count(lettre)
        n = occurence
        if n%2 == 0:
            indice=n//2
        else:
            indice = n*2
        ordre = alpha.find(lettre)
        ordre2 = (ordre + indice) % len(alpha)
        nouvelle_lettre = alpha[ordre2]
        chiffre += nouvelle_lettre
    return chiffre
print(chiffrer("HAPPY"))
```

# Programme 3

Un des plus anciens systèmes de cryptographie (aisément déchiffrable) consiste à décaler les lettres d’un message pour le rendre illisible. Ainsi, les __A__ deviennent des __B__, les __B__ des __C__, etc. et les __Z__ deviennent des __A__.

Exemple :

Si le message en clair __message__ = ABCCZABEY alors le message chiffré est BCDDABCFZ

## Question

Ecrire un programme qui, à partir d'une chaine __message__, la chiffre dans un message __chiffre__ selon ce principe.

```python
def chiffrer(message) :
    chiffre = ""
    alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    for lettre in message :
        ordre = alpha.find(lettre)
        ordre2 = (ordre + 1) % len(alpha)
        nouvelle_lettre = alpha[ordre2]
        chiffre += nouvelle_lettre
    return chiffre
print(chiffrer("ABCCZABEY"))
```

# Programme 4

Une amélioration (relative) du principe utilisé dans le programme 3 consiste à opérer avec un décalage non pas de 1, mais d’un nombre quelconque de lettres (chiffre de césar). Ainsi, par exemple, si on choisit un décalage de *3*, les __A__ deviennent des __D__, les __B__ des __E__, etc. et les __Z__ deviennent __C__

## Question

Ercire un programme sur le même principe que le précédent, mais qui demande en plus le décalage à utiliser.

```python
def chiffrer_cesar(message, decalage):
    chiffre = ""
    alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    
    for lettre in message :
        ordre = alpha.find(lettre)
        ordre2 = (ordre + decalage) % len(alpha)
        nouvelle_lettre = alpha[ordre2]
        chiffre += nouvelle_lettre
    return chiffre
print(chiffrer_cesar("ABCDEFGH", 3))
```

# Programme 5

Une technique de chiffrement consiste à opérer non pas avec un décalage systématique, mais par une substitution aléatoire. 

Pour cela,on utilise un *alphabet-clé*, dans lequel les lettres se succèdent de manière désordonnée.

Par exemple, __HYLUJPVREAKBNDOFSQZCWMGITX__, c’est cette clé qui va servir ensuite à chiffrer le message. Selon cet exemple, les __A__ deviennent des __H__, les __B__ des __Y__, les __C__ des __L__, etc.

Exemple :

Si le message en clair __message__ = ABCDEFZ alors le message chiffré est HYLUJPX

## Question

Ecrire un programme qui effectue ce chiffrement sur un message __message__ et produit le chiffré __chiffre__. L'*alphabet-clé* sera directement présente dans le programme.

```python
def chiffrer_alea(message):
    chiffre = ""
    alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    alpha_clé = 'HYLUJPVREAKBNDOFSQZCWMGITX'
    for lettre in message :
        ordre = alpha.find(lettre)
        nouvelle_lettre = alpha_clé[ordre]
        chiffre += nouvelle_lettre
    return chiffre
print(chiffrer_alea("ABCDEFZ"))
```
